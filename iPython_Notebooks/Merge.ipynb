***

```python
#REad in ncaa data
ncaa = pd.read_csv("NCAA_clean_Final.csv")

ncaa = ncaa.sort_values("Season").drop_duplicates(subset="Player", keep = "last")
```

```python
#read in second dataset for NCAA clean
ncaa_2 = pd.read_csv("NCAA_clean2.csv")
ncaa_2
```

```python
cols2 = ["Player", "eFG%", "WS"]
ncaa_sub = ncaa_2[cols2]
ncaa_sub.head()
```

```python
#Get WS and effective field goal percentage

first_merge = ncaa.merge(ncaa_sub, on = "Player", how="left")
```

```python
#Drop unnamed column

first_merge = first_merge[first_merge.Season.notna()].drop_duplicates("Player", keep = "last")
```

```python
first_merge["WS"] = first_merge.apply(lambda row: 0.56 if (np.isnan(row["WS"])) & (row["PER"] >=20) else row["WS"], axis = 1)
```

```python
first_merge["eFG%"] = first_merge.apply(lambda row: 0.39 if (np.isnan(row["eFG%"])) & (row["PER"] >=20) else row["eFG%"], axis = 1)
```

```python
first_merge = first_merge[first_merge.WS.notna()]
```

```python
#make sure duplicates are dropped

first_merge = first_merge.drop_duplicates(subset="Player", keep="last")
```
