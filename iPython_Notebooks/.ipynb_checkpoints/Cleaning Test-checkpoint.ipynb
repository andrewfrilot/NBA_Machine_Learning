```python
#--Dependencies--#

#-- Data Cleaning Libraries:
import pandas as pd
import numpy as np

#-- Data Visualization Libraries:
from matplotlib import pyplot as plt
import seaborn as sns #--just in case we need it, probably won't


#-- Web Scraping Libraries:
import os
import time
import requests
from bs4 import BeautifulSoup as bs
from splinter import Browser
```

```python
#Read CSV
raw_df = pd.read_csv("players_birth_raw.csv")
```

```python
#Create function to take out year

def year_func(row):
    year = row[-5: -1]
    
    return year

def name_func(row):
    name = row[0:-6]
    
    return name
```

```python
raw_df.columns
```

```python
raw_df["Year"]= raw_df["1-Jan"].apply(lambda row : year_func(row))
```

```python
raw_df["1-Jan"] = raw_df["1-Jan"].apply(lambda row: name_func(row))

```

```python
raw_df = raw_df.rename(columns={"1-Jan": "Player"})
```
