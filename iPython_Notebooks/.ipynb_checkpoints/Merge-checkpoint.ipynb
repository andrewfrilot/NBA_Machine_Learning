## Manipulating Data for Model

### NCAA data merged with WS and eFG data. Merged next with awards data

```python
#read in draft data
draft_data = pd.read_excel("../Draft2.xlsx")
```

```python
draft_data.head()
```

```python
drafty = draft_data[["Year", "Lg", "Rd", "Pk", "Tm", "Player"]]
```

***

```python
#REad in ncaa data
ncaa = pd.read_csv("NCAA_clean_Final.csv")

ncaa = ncaa.sort_values("Season").drop_duplicates(subset="Player", keep = "last")
```

```python
#read in second dataset for NCAA clean
ncaa_2 = pd.read_csv("NCAA_clean2.csv")
ncaa_2
```

```python
cols2 = ["Player", "eFG%", "WS"]
ncaa_sub = ncaa_2[cols2]
ncaa_sub.head()
```

```python
#Get WS and effective field goal percentage

first_merge = ncaa.merge(ncaa_sub, on = "Player", how="left")
```

```python
#Drop unnamed column

first_merge = first_merge[first_merge.Season.notna()].drop_duplicates("Player", keep = "last")
```

```python
first_merge["WS"] = first_merge.apply(lambda row: 0.56 if (np.isnan(row["WS"])) & (row["PER"] >=20) else row["WS"], axis = 1)
```

```python
first_merge["eFG%"] = first_merge.apply(lambda row: 0.39 if (np.isnan(row["eFG%"])) & (row["PER"] >=20) else row["eFG%"], axis = 1)
```

```python
first_merge = first_merge[first_merge.WS.notna()]
```

```python
#make sure duplicates are dropped

first_merge = first_merge.drop_duplicates(subset="Player", keep="last")
```

```python
first_merge[first_merge.Player == "Zion Williamson"]
```

***

```python
#read in awards data

awards = pd.read_csv(r"..\NCAA_Player_Honors.csv")
```

```python
awards.head()
```

```python
#Convert Seasons into a single year value for awards

def normalize_year(y):

    if y == "1999-00":
        year = "2000"
    elif y == "2009-10":
        year = "2010"
    elif y == "2019-20":
        year = "2020"
    else:
        year = int(y[0:3] + y[-1])
    
    return year

```

```python
awards.Year = awards.Year.apply(lambda x: normalize_year(x))
```

```python
player_list = awards.Player.str.split("(")
```

```python
players = []

for player in player_list:
    players.append(player[0])
```

```python
awards.Player = players
```

```python
#now change to award to 1's

awards.Award = 1
```

```python
awards.head()
```

```python
awards_agg = awards.groupby("Player").agg(np.sum)
```

```python
awards_agg = awards_agg.reset_index()
```

```python
awards_agg.Player[0]
```

```python
#left merge on to first_merge

test_merge = first_merge.merge(awards_agg, on = "Player", how="left")
```

```python
test_merge.Award.fillna(0, inplace = True)
```

```python
test_merge
```

```python
#Send to csv

first_merge.to_csv("ncaa_avg_stats.csv")
```

```python
#merge draft data with ncaa data

final_merge = test_merge.merge(drafty, on = "Player", how="left")
```
